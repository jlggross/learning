# Methods

* Are functions that belong to a certain object
* The method is accessed by typing a dot ('.')

````
# Methods for list

l = list() # create empty list
print(l)

# Add value at the end of the list
# First element is 0
l.append("me") 
l.append("a")
l.append("is")
l.append("me")
print(l)

# Return index of element
print(l.index("a")) 
try:
	print(l.index("bla")) # Gives error, because element does not exist in list
except:
	print("Error - Element does not exist in list")

# Count number of elements
print(l.count("me"))

# Remove the element
l.remove("a")
print(l)
````

````
# Methods for string

s = "my name is Ali Baba"

# First occurence of element
print(s.index("a"))

# Set first element to capital size
print(s.capitalize())

# Replace all occurencies of the element
print(s.replace("a", "XX"))
````
