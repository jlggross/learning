# Random Numbers

```
# Random Generators
import numpy as np

print(np.random.rand())	# Pseudo-random numbers

np.random.seed(123)	# Starting from a seed
print(np.random.rand())
```

```
# Coin Toss
import numpy as np

#np.random.seed(123)	# Starting from a seed

for i in range (0,10):
    coin = np.random.randint(0,2)	# Randomly generate 0 or 1
    if coin == 0:
        print(i, "heads")
    else:
        print(i, "tails")
```
