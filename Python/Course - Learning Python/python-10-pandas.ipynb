# Pandas package

* Creating a DataFrame
* Reading data from file
* Accessing rows and columns

```
# DataFrame
import pandas as pd

names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]
my_dict = { 'country':names, 'drives_right':dr, 'cars_per_cap':cpc }

cars = pd.DataFrame(my_dict) 
print(cars)

row_labels = ['US', 'AUS', 'JPN', 'IN', 'RU', 'MOR', 'EG']
cars.index = row_labels # Specify row labels of cars
print("\n", cars)
```

```
# Read from CSV file
cars = pd.read_csv('cars.csv', index_col=0) # lê do arquivo e diz que o índice do DataGrame é a coluna 0 do arquivo
```

## Accessing values from rows and columns

```
# Column Access []
print(cars[['country', 'cars_per_cap']]) # retorna apenas as colunas selecionadas do DataFrame

# Row Access []
print('\n', cars[1:4]) #seleciona as linhas correspondentes 1, 2 e 3
```

```
# Row Access loc
print(cars.loc[['US']]) # retorna toda informações sobre esse índice
print('\n', cars.loc[['US', 'JPN', 'IN']])
print('\n', cars.loc[['US', 'JPN', 'IN'], ['country', 'cars_per_cap']])

# Row Access iloc
print('\n', cars.iloc[[0]])
print('\n', cars.iloc[[0, 2, 3]])
print('\n', cars.iloc[[0, 2, 3], [0, 2]])
```
