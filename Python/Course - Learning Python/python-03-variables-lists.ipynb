```
savings = 100.0     # float
height = "190"      # string
meter = 10          # int

str(savings) 	# convert to string
int(height)		# convert to integer
float(meter)	# convert to float
bool(meter)		# convert to bool

result = 100 * 1.10 ** 7
print("I started with $" + str(savings) + " and now have $" + str(result))
```

## Working with lists

```
# Declaring lists

fam = [1.8, 1.9, 2.4, 1.76] # lista simples

fam2 = [["dad", 1.86],
		["mom", 1.74],
		["love", 1.74],
		["me", 1.89]]		# lista de listas - matriz
		
print(type(fam)) 	# list
print(type(fam2))	# list
```

```
# Accessing values

print(fam[0]) 	# retorna 1.8
print(fam[2])	# retorna 2.4
print(fam[-1])	# último elemento da lista = 1.76
print(fam[3])	# retorna 1.76
```

```
# List slicing

"""
my_list[start:end]  # The start index will be included, 
					# while the end index is not.
"""

print(fam[1:3]) 	# retorna sublista de fam com elementos [1.9, 2.4]
			        # o final da seleção não é selecionado

print(fam[:3])		# retorna sublista [1.8, 1.9, 2.4]
print(fam[2:])		# retorna sublista [2.4, 1.76]

print(fam2[-1][0]) # retorna "me"

fam[0:2] = [2.0, 2.2]
print(fam) # [2.0, 2.2, 2.4, 1.76]
```

```
# Adding and removing elements

l = fam + [3.0, 4.0] # fam fica com [2.0, 2.2, 2.4, 1.76, 3.0, 4.0]
print(l)

del(l[-1]) 	 # fam fica com [2.0, 2.2, 2.4, 1.76, 3.0]
print(l)
```

```
# List reference and copy

x = ["a", "b", "c"]
y = x 		# diferentes referências à mesma lista

print(x)
print(y)

y[1] = "z"	# x = ["a", "z", "c"], y = ["a", "z", "c"]

print(x)
print(y)
```

```
# List copy

x = ["a", "b", "c"]
y = list(x) # cria uma nova lista para y

print(x)
print(y)

y[1] = "z"	# x = ["a", "b", "c"], y = ["a", "z", "c"]

print(x)
print(y)

y = x[:] # Subsetting a list always returns a new list. Its is one way to copy a list
print(y)
```
