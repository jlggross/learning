```
# Package import
import numpy as np

a = np.array([21, 22, 23, 24, 25])
print(a[a > 23]) # retorna [24, 25]
```

```
import numpy as np
a = np.array([21, 22, 23, 24, 25])
my_bools = a > 23	# cria lista de bools com False e True
print(a[my_bools]) 	# imprime apenas elementos True, ou seja,
					# > 23
```

```
np_2d = np.array([	[1.74, 1.64, 1.86, 1.89],
					[83, 74, 115, 79] ])
print(np_2d.shape) # retorna a forma da estrutura
print(np_2d[0][2]) # retorna 1.86
print(np_2d[0,2]) 	# retorna 1.86
print(np_2d[:,1:3])	# retorna as duas linhas, porém apenas as colunas 1 e 2

np_1d = np_2d[:,0] # retorna todas as linhas, mas apenas a coluna 0
print(np_1d)
```

## Numpy Statistics

```
print(np_2d.mean()) # média de um conjunto de valores
```
