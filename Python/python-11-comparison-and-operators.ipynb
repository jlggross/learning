# Comparison Operators

````
import numpy as np

print(2 < 3) 	# True
print(2 == 3) 	# False
print(2 <= 3)	# True
print(2 != 3) 	# True
print("carl" < "chris") # True

try:
    print(3 < "chris")	# Error
except:
    print("Error!!!")

bmi = np.array([21.852, 20.975, 21.75, 24.747, 21.441])
print(bmi > 23) # array([False, False, False, True, False])
````

# Boolean Operators

* For Numpy Arrays: \
```logical_and()```\
```logical_or()```\
```logical_not()```

````
print(True and False)	# False
print(False or True)	# True
print(not True)		# False
````

````
import numpy as np

bmi = np.array([21.852, 20.975, 21.75, 24.747, 21.441])
print(bmi > 23) # array([False, False, False, True, False])

try:
    print(bmi > 23 and bmi < 22)	# Error
except:
    print("Error!!!")

print(np.logical_and(bmi > 21, bmi < 22)) # array([True, False, True, False, True])

print(bmi[np.logical_and(bmi > 21, bmi < 22)]) # array([21.852, 21.75, 21.441])
````
