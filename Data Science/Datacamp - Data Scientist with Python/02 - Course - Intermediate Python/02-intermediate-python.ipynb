# Chapter 2 - Dictionaries & Pandas

Learn about the dictionary, an alternative to the Python list, and the pandas DataFrame, the de facto standard to work with tabular data in Python. You will get hands-on practice with creating and manipulating datasets, and you’ll learn how to access the information you need from these data structures.

## Dictionaries

* Convenient and intuitive way to associate values
* Each value is accessed by a unique key
  * In lists we use `indexes`
* The keys in a dictionary have to be "immutable" objects


### Creating a dictionary

A simple recipe for creating a dictionary:

```
my_dict = {
   "key1":"value1",
   "key2":"value2",
}
```

````
world = {"afghanistan":30.55, 
         "albania":2.77, 
         "algeria":39.21}
print(world)
print(world["albania"])
````

````
# With immutabe keys
my_dict = {0:"hello", True:"dear", "two":"world"}
````

````
# List as keys gives error
another_dict = {["just", "to", "test"]: "value"}
````

### Access dictionary

If the keys of a dictionary are chosen wisely, accessing the values in a dictionary is easy and intuitive.

````
# Definition of dictionary
europe = {'spain':'madrid', 
          'france':'paris', 
          'germany':'berlin', 
          'norway':'oslo' }

# Print out the keys in europe
print(europe.keys())

# Print out value that belongs to key 'norway'
print(europe['norway'])
````

### Creating and deleting elements

````
world = {"afghanistan":30.55, 
         "albania":2.77, 
         "algeria":39.21}
print(world)
````

````
# New element
world["brazil"] = 201.30
print(world)
````

````
# Remove element
del(world["albania"])
print(world)
````

### Using `in`

````
world = {"afghanistan":30.55, "albania":2.77, "algeria":39.21}
print("russia" in world)
print("algeria" in world)
````

### Access value using `loc`

* Square brackets have limited functionality
* Ideally
  * 2D Numpy arrays
  * `my_array[rows, columns]`
* pandas
  * `loc` (label-based)
  * `iloc` (integer position-based)

````
# Find row by index label - Return Series
print(brics.loc["RU"])
````

````
# Find row by index label - Return DataFrame
print(brics.loc[["RU"]])
````

````
print(brics.loc[["RU", "IN", "CH"]])
````

````
print(brics.loc[["RU", "IN", "CH"], ["country", "capital"]])
````

````
print(brics.loc[:, ["country", "capital"]])
````

### Square brackets vs. `loc`

* Square brackets
  * Column access `brics[["country", "capital"]]`
  * Row access: only through slicing `brics[1:4]`
* loc (label-based)
  * Column access `brics.loc[:, ["country", "capital"]]`
  * Row access `brics.loc[["RU", "IN", "CH"]]`
  * Row & Column access 

  ```
  brics.loc[["RU", "IN", "CH"], ["country", "capital"]]
  ```

### Access value using `iloc`

# Chapter 4 - Loops

There are several techniques you can use to repeatedly execute Python code. While loops are like repeated if statements, the for loop iterates over all kinds of data structures. Learn all about them in this chapter.

## While

* Repeat until condition is met

```
while condition :
    expression
```

````
error = 50.0

while error > 1:
    error = error / 2
    print(error)
````

## For loop

* For each element in the sequence, execute expression

```
for var in seq :
    expression
```

````
fam = [1.73, 1.68, 1.71, 1.89]
for height in fam :
    print(height)
````

### Enumerate

````
fam = [1.73, 1.68, 1.71, 1.89]
for index, height in enumerate(fam) :
    print(index, height)
````

## Random Numbers

````
import numpy as np
np.random.seed(123) # Choosing a seed ensures "reproducibility"
np.random.rand()
````

````
# Coin toss
import numpy as np
np.random.seed(123)
coin = np.random.randint(0,2) # Randomly generate 0 or 1

if coin == 0:
    print("heads")
else:
    print("tails")
````

````
# Random Walk
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123)
final_tails = []

for x in range(10000) : # <--
    tails = [0]
    for x in range(10) :
        coin = np.random.randint(0, 2)
        tails.append(tails[x] + coin)
    final_tails.append(tails[-1])
plt.hist(final_tails, bins = 10)
plt.show()
````
