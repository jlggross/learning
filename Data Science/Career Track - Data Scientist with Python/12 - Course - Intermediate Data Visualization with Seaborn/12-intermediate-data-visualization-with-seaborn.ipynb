## Using the distribution plot

### Creating a histogram

* Distplot function ```distplot()``` has multiple optional arguments.
* When used it will return a Kernel Density Estimate (KDE) by default.
  * The KDE helps to smooth the distribution and is a useful way to look at the data. 
* In order to plot a simple histogram, you can disable the kde and
specify the number of bins to use:

    ```sns.distplot(df['alcohol'], kde=False, bins=10)```

````
# Distribution plot without kde off
import seaborn as sns
import pandas as pd
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Career Track - Data Scientist with Python/12 - Course - Intermediate Data Visualization with Seaborn/datasets/wines.csv')
sns.distplot(df['Flavanoids'], kde=False, bins=10)
````

### Defining a color for a plot

### Palettes

* Seaborn uses the ```set_palette()``` function to define a palette.
* ```sns.palettes.SEABORN_PALETTES``` has a list of different color palettes one can use.

* To display a palette the function ```sns.palplot()``` can be used
* Using ```sns.color_palette()``` returns the current palette
    

````
for i, p in enumerate(sns.palettes.SEABORN_PALETTES, start=0):
    if i % 2 == 1:
        continue    
    sns.set_palette(p)
    #sns.distplot(wines['Flavanoids'])
    sns.palplot(sns.color_palette())    
````

### Defining Custom Palettes

````
# Setting and displaying custom color palettes
sns.palplot(sns.color_palette('Paired', 20)) # Circular colors - when the data is not ordered
sns.palplot(sns.color_palette('Blues', 12)) # Sequential colors - when the data has a consistent range from high to low
sns.palplot(sns.color_palette('BrBG', 8)) # Diverging colors - when both the low and high values are interesting
````

## Regression Plots

### Bicycle Dataset

* Aggregated bicycle sharing data in Washington DC
* Data includes:
  * Rental amounts
  * Weather information
  * Calendar information
* Can we predict rental amounts?

````
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

filepath = '/content/drive/My Drive/Colab Notebooks/Career Track - Data Scientist with Python/12 - Course - Intermediate Data Visualization with Seaborn/datasets/'
bikes = pd.read_csv(filepath + 'bike_share.csv')
bikes.head(3)
````

### Plotting with `regplot()`

* ```regplot()```: Regression plot
  * ```order``` parameter is used to set the order of the polynomial regression
  * ```x_estimator``` parameter can be useful for highlighting trends
    * Provides simplified view

````
fig, (ax0, ax1, ax2) = plt.subplots(nrows=1,ncols=3, sharey=True, figsize=(12,4))

sns.regplot(data=bikes, x='temp', y='total_rentals', marker='+', ax=ax0)

# Polynomial regression using the 'order' parameter
sns.regplot(data=bikes, x='temp', y='total_rentals', order=2, color='r', ax=ax1)
sns.regplot(data=bikes, x='temp', y='total_rentals', order=2, color='g', 
            fit_reg=False, ax=ax2)
````

### Evaluating regression with `residplot()`

* ```residplot()```: Residual plot
  * A residual plot is useful for evaluating the fit of a model

````
fig, (ax0, ax1) = plt.subplots(nrows=1,ncols=2, sharey=True, figsize=(10,3))

sns.residplot(data=bikes, x='temp', y='total_rentals', color='b', ax=ax0)

# Polynomial regression with 'order'
sns.residplot(data=bikes, x='temp', y='total_rentals', order=2, color='y', ax=ax1)
````

### Polynomial regression

* Seaborn supports polynomial regression using the `order` parameter

````
fig, (ax0, ax1) = plt.subplots(nrows=1,ncols=2, sharey=True, figsize=(10,3))
sns.regplot(data=bikes, x='temp', y='total_rentals', 
            x_jitter=.1, order=2, ax=ax0)
sns.regplot(data=bikes, x='temp', y='total_rentals', order=2, ax=ax1)
````

### `residplot` with polynomial regression

````
sns.residplot(data=bikes, x='temp',y='total_rentals', order=2)
````

### Categorical values

````
# Easier to see the distribution of rentals each month
sns.regplot(data=bikes, x='mnth', y='total_rentals', x_jitter=.1, order=2)
````

### Estimators

* In some cases, an `x_estimator` can be useful for highlighting trends

````
# Estimator
import numpy as np

fig, (ax0, ax1, ax2) = plt.subplots(nrows=1,ncols=3, sharey=True, figsize=(10,3))
sns.regplot(data=bikes, x='mnth', y='total_rentals',
            x_estimator=np.mean, order=2, ax=ax0)
sns.regplot(data=bikes, x='mnth', y='total_rentals',
            x_estimator=np.median, order=2, ax=ax1)
sns.regplot(data=bikes, x='mnth', y='total_rentals',
            x_estimator=np.max, order=2, ax=ax2)
````

### Binning the data

* `x_bins` can be used to divide the data into discrete bins
* The regression line is still t against allthe data

````
# Selecting number of bins
fig, (ax0, ax1, ax2) = plt.subplots(nrows=1,ncols=3, sharey=True, figsize=(10,3))
sns.regplot(data=bikes, x='temp', y='total_rentals', x_bins=4, ax=ax0)
sns.regplot(data=bikes, x='temp', y='total_rentals', x_bins=8, ax=ax1)
sns.regplot(data=bikes, x='temp', y='total_rentals', x_bins=12, order=3, ax=ax2)
````

## Matrix Plots

### Getting data in the rightformat

* Seaborn's `heatmap()` function requires data to be in a grid format
* pandas `crosstab()` is frequently used to manipulate the data
* Matrix plots can be used to quicky see trends in a dataset


````
# Create table to be used in the heatmap
bikes_crosstab = pd.crosstab(bikes['mnth'], bikes['weekday'],
            values=bikes['total_rentals'],aggfunc='mean').round(0)
print(bikes_crosstab)
````

### Build a heatmap

````
sns.heatmap(pd.crosstab(bikes['mnth'], bikes['weekday'], 
                        values=bikes['total_rentals'], 
                        aggfunc='mean'))
````

## Customize a heatmap
* ```annot=True```: Turn on annotations in the indiividual cells.
* ```fmt='d'```: Uses formating to show values as integers.
* ```fmt='.1f'```: Uses formating to show values as floats with one decimal number.
* ```cmap```: Parameter used to change the shading used in the plot.
* ```cbar=False```: The color bar is not displayed.
* ```linewidths=0.5```: Puts a small space between cells, so that the values are simpler to view.
* ```center=1000```: Center the color scheme of the heatmap to the value 1000.

````
sns.heatmap(bikes_crosstab, annot=True, fmt='.0f',
            cmap='YlGnBu', cbar=False, linewidths=.5)
````

### Centering a heatmap

* Seaborn support centering the heatmap colors on a specic value

````
sns.heatmap(bikes_crosstab, annot=True, fmt= '.0f',
            cmap= 'YlGnBu', cbar=True, center=bikes_crosstab.loc[9, 6])
````

### Plotting a correlation matrix
* Pandas ```corr``` function calculates correlations between columns in a dataframe
* The output can be converted to a heatmap with seaborn

    ```sns.heatmap(df.corr())```

````
sns.heatmap(bikes.corr())
````

# Chapter 4 - Creating Plots on Data Aware Grids

Using Seaborn to draw multiple plots in a single figure.

## Using FacetGrid, factorplot and lmplot

* Multiple smaller plots can build a larger visualization that can help identy trends in data with manu variables.
* The concept of small multiples is useful for analysing data with many variables.
* The idea is that you can quickly identify trends in data by comparing multiple plots side by side using the same scales and axes.
  * This process is called **faceting**.


### Tidy data

* Seaborn's grid plots require data in 'tidy format'
* One observation per row of data

### Facegrid

* The ```FacetGrid``` is foundational for many data-aware grids.
* A data-aware grid allows you to create a series of small plots that can be useful for understanding complex data relationships.
* It allows the user to control how data is distributed across columns,
rows and hue.
* Once a `FacetGrid` is created, the plot type must be mapped to the grid
* When building a FacetGrid, there are two steps:
  1. Create a ```FacetGrid``` object with columns, rows, or hue.
  2. Map individual plots to the grid.

### FacetGrid Categorical Example

````
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_style('white')
filepath = '/content/drive/My Drive/Colab Notebooks/Career Track - Data Scientist with Python/12 - Course - Intermediate Data Visualization with Seaborn/datasets/'
college = pd.read_csv(filepath + '2018 College Scorecard Tuition.csv', index_col=0)
college.head(3)
````

````
g = sns.FacetGrid(college, col='HIGHDEG', col_order=[1, 2, 3, 4])
g.map(sns.boxplot, 'Tuition')
````

* The `factorplot` is a simpler way to use a `FacetGrid` for categorical data
* Combines the facetting and mapping process into 1 function

````
sns.factorplot(x='Tuition', data=college, col='HIGHDEG', kind='box')
````

## Using JointGrid and jointplot

### JointGrid() Overview

* A ```JointGrid()``` allows us to compare the distribution of data beteween two variables.
  * ```JointGrid()``` combines univariate plots such as histograms, rug plots and kde plots with bivariate plots such as scatter and regression plots.
* Takes a small number of inputs and creates an insightful visualization of the data.
* Layout basics of a ```JointGrid()```:
  * The input to ```JointGrid()``` is an X and Y variable. 
  * The center of the plot contains a plot type, for example, a scatter plot of these two variables. 
  * The plots along the x and y-axis show the distribution of the data for each variable.
  * This plot can be configured by specifying the type of joint plots as well as the marginal plots. 


````
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('white')
filepath = '/content/drive/My Drive/Colab Notebooks/Career Track - Data Scientist with Python/12 - Course - Intermediate Data Visualization with Seaborn/datasets/'
college = pd.read_csv(filepath + '2018 College Scorecard Tuition.csv', index_col=0)
college.head(3)
````

### Basic JointGrid

````
# Basic JointGrid
g = sns.JointGrid(data=college, x='Tuition', y='ADM_RATE_ALL')
g.plot(sns.regplot, sns.distplot)
````

### Customizing JointGrid()

````
# Customizing JointGrid()
import numpy as np
g = sns.JointGrid(data=college, x='Tuition', y='ADM_RATE_ALL')

g = g.plot_joint(sns.scatterplot, color='green', alpha=0.6)
_ = g.ax_marg_x.hist(college['Tuition'], color='red', alpha=.6,
                     bins=np.arange(0, 50000, 5000))
_ = g.ax_marg_y.hist(college['ADM_RATE_ALL'], color='blue', alpha=.6,
                     orientation='horizontal',
                     bins=10)
````

### Advanced JointGrid

````
# Advanced JoinGrid
g = sns.JointGrid(data=college, x='Tuition', y='ADM_RATE_ALL', 
                  xlim=(0, 45000), ylim=(0.2, 1.2))

#color_map_palette = sns.dark_palette('palegreen', as_cmap=True)
color_map_palette = sns.light_palette('green', as_cmap=True)
#color_map_palette = 'Blues_d'
#color_map_palette = 'Reds_d'
#color_map_palette = 'Purples_d'
g = g.plot_joint(sns.kdeplot, 
                 cmap=color_map_palette,
                 n_levels=8)

#g = g.plot_marginals(sns.kdeplot, c='g', shade=True)
g = g.plot_marginals(sns.distplot, color='g')
````

````
# Advanced JoinGrid
g = sns.JointGrid(data=college, x='Tuition', y='ADM_RATE_ALL', 
                  xlim=(0, 45000), ylim=(0.2, 1.2))
color_map_palette = sns.light_palette('blue', as_cmap=True)
g = g.plot_joint(sns.kdeplot, 
                 cmap=color_map_palette,
                 n_levels=8,
                 shade=True)
g = g.plot_marginals(sns.kdeplot, color='blue', shade=True)
````

### ```jointplot()```
* It is easier to use but provides fewer customization capabilities. 
* The jointplot is a convenience wrapper around many of the ```JointGrid``` functions.

````
import scipy.stats as stats
g = sns.jointplot(data=college, x='Tuition', y='ADM_RATE_ALL', kind='hex')
g.annotate(stats.pearsonr)
````

### Customizing a jointplot

````
# Customizing a jointplot - with overlay plots
import scipy.stats as stats
g = (sns.jointplot(x='Tuition', y='ADM_RATE_ALL', 
                   kind='scatter', 
                   xlim=(0, 25000), 
                   marginal_kws=dict(bins=15, rug=True), 
                   data=college.query("UG < 2500 & Ownership == 'Public'"))
    .plot_joint(sns.kdeplot))
g.annotate(stats.pearsonr)
````

## Selecting Seaborn Plots

### Univariate Distribution Analysis
* ```distplot()``` is the best place to start for this analysis
  * Is the best place to start when trying to do distribution analysis with Seaborn.
  * Combines many of the features of the ```rugplot()```, ```kdeplot()```, and matplotlib histogram into a single function. 
  * Can be used to generated kdeplot plot. 
* ```rugplot()``` and ```kdeplot()``` can be useful alternatives

### Regression Analysis
A regression plot is an example of a plot that shows the relationship between two variables. 

* matplotlib's ```scatter()``` plot is a very simple method to compare the interaction of two variables on the x and y-axis. 
* ```lmplot()``` performs regression analysis and supports facetting.
  * The ```lmplot()``` combines many of these features of the underlying ```regplot()``` and ```residplot()``` in addition to the ability to plot the data on a ```FacetGrid()```. 
  * Is the best function, in many instances, to use for determining linear relationships between data.

### Categorical Plots
Explore data with the categorical plots and facet with

* ```boxplot()``` or ```violinplot()``` can be used to examine the distribution of the variables. 
  * Can be combined with the statistical estimation plots such as the point, bar, or countplot.
* To facet the data across rows or columns we use ```factorplot()```.

### pairplot() and jointplot()
* ```PairGrid()``` and ```JointGrid()```: useful visualizations after some preliminary analysis of regressions or distributions of the data has already been done. 
  * Perform regression analysis with ```lmplot```.
  * Analyze distributions with ```distplot```.
* After being familiar with the data, the ```pairplot()``` and ```jointplot()``` can be used to understand how two or more variables interact with each other.
