### Other configurations for scatter plot

* `hue` : define color of points according to the specified column.

```
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x="total_bill", y="tip", 
                data=tips, hue="smoker")
plt.show()
```

* `hue_order` : define de order in which the colors specified in `hue` will be shown.

```
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x="total_bill", y="tip", 
                data=tips, hue="smoker",
                hue_order=["Yes", "No"])
plt.show()
```

* `palette` : specify a palette of colors for the different group of points in `hue`

````
# Specifying hue colors
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tips = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Career Track - Data Scientist with Python/09 - Course - Introduction to Data Visualization with Seaborn/datasets/tips.csv")
hue_colors = {"Yes": "black", 
              "No": "red"}
sns.scatterplot(x="total_bill", y="tip", data=tips, 
                hue="smoker", hue_order=["Yes", "No"],
                palette=hue_colors)
plt.show()
````

### Introducing `relplot()`

* Create "relational plots": scatter plots or line plots
* Why use `relplot()` instead of `scatterplot()` ?
  * `relplot()` lets you create subplots in a single figure

### scatterplot() vs. relplot()

````
# Using scatterplot()
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x="total_bill", y="tip",data=tips)
plt.show()
````

````
# Using relplot()
import seaborn as sns
import matplotlib.pyplot as plt
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter")
plt.show()
````

## Creating a box plot

### What is a box plot?
* Shows the distribution of quantitative data
* Show median, spread, skewness, outliers and confidence intervals
* Facilitates comparisons between groups
* It's encoraged to learn more about statistical analysis to understand the fundamentals of all the information that is been shown in the box plot.
  * Also, an statictical analysis course will help you edit the box plot accordingly to your needs.

### Usage
* Use of ```catplot```, with ```kind="box"```
* ```sym=""```: Ommit the outliers from the plot
* ```whis=2.0```: Make whiskers extend to 2.0 * IQR
* ```whis=[5, 95]```: Show the 5th and 95th percentiles
* ```whis=[0, 100]```: Show min and max values

### Creating a box plot with ordered categories

````
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Creating dataset
student_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Career Track - Data Scientist with Python/09 - Course - Introduction to Data Visualization with Seaborn/datasets/student_data.csv", index_col=0)

# Preparing box plot
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]
sns.catplot(x="study_time", y="G3", 
            kind="box", 
            data=student_data, 
            order=study_time_order, 
            sym="") # Remove outliers
plt.show()
````

### Omitting the outliers using `sym`

````
# Create a box plot with subgroups and omit the outliers
sns.catplot(x="internet", y="G3", 
            kind="box", 
            data=student_data, 
            hue="location",
            sym=""   # Remove outliers
            #, whis=0.5 # Change whiskers from 1.5 to 0.5*IQR
            #, whis=[0, 100] # Change whiskers to min (0) and máx (100)
            ) 
plt.show()
````

### Changing the whiskers using `whis`

* By default,the whiskers extend to 1.5 * the interquartile range
* Make them extend to 2.0 * IQR: `whis=2.0`
* Show the 5th and 95th percentiles: `whis=[5, 95]`
* Show min and max values: `whis=[0, 100]`


````
# Create a box plot with subgroups and omit the outliers
sns.catplot(x="internet", y="G3", 
            kind="box", 
            data=student_data, 
            hue="location",
            whis=[0, 100]) # Change whiskers to min (0) and máx (100) 
plt.show()
````

## Point plots

### What are point plots?
* Point plots have similiar information as box plots
* Points show mean of quantitative variable
* Vertical lines show 95% confidence intervals
 * Confidence intervals: show the level of uncertainty about the mean stimates.
 * Assuming that the data is a random sample of the population, we can be 95% sure that the true population mean in each group lies within the confidence interval shown.

### Usage
* ```catplot()``` with ```kind="point"```
* ```join=False```: Does not unite the points on the plot

### Point plots vs. line plots vs. bar plot

All three have:
* Mean of quantitative variable
* 95% condence intervals for the mean

Differences:
* **Line plot**: has **quantitative** variable on x-axis (usually time).
* **Point plot**: has **categorical** variable on x-axis.

### Creating a point plot

````
# Family relationship vs. absences
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

student_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Career Track - Data Scientist with Python/09 - Course - Introduction to Data Visualization with Seaborn/datasets/student_data.csv", index_col=0)
sns.catplot(x="famrel", y="absences", 
            kind="point", 
            hue="school",
            data=student_data)
plt.show()
````

### Creating a point plot with subgroups

* `hue` : When specified create subgroups.
* `estimator` : white `np.median` displays the median.

````
import numpy as np

# Creating a point plot with subgroups
sns.catplot(x="romantic", y="absences", 
            kind="point", 
            data=student_data,
            hue="school", # Make plots with different colors
            ci=None, # Confidence interval is off
            estimator=np.median) # Display the median number of absences instead of the average
plt.show()
````

### Disconnecting the points

````
sns.catplot(x="famrel", y="absences", 
            kind="point", 
            hue="school",
            data=student_data,
            join=False)
plt.show()
````

### Customizing the confidence intervals

````
sns.catplot(x="famrel", y="absences", 
            kind="point", 
            hue="school",
            data=student_data,
            capsize=0.2)
plt.show()
````

### Turning off confidence intervals

````
sns.catplot(x="famrel", y="absences", 
            kind="point", 
            hue="school",
            data=student_data,
            ci=None)
plt.show()
````
