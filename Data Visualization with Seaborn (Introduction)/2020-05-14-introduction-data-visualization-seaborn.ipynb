## Overview

* Powerful data visualization tool built on top of Matplotlib
* Works very well with DataFrames from Pandas library
* Can generate attractive and informative visualization with just a few lines of code.
* Helpful for data exploration and to communicate results after data analysis has been done.

## Agenda
* Importing ```seaborn```
* Relational plots
  * Scatter plots, line plots
* Categorical plots
  * Count plots, box plots, point plots
* Display information
  * Using ```hue``` to display information in different colors
  * Using ```rol``` and ```col``` in ```relplot()``` to make multiples subplots of subgroups of data
* Customization
  * Change the background: ```sns.set_style()```
  * Change the main element colors: ```sns.set_palette()```
  * Change the scale: ```sns.set_context()```
  * Adding title
   * Differences of ```FacetGrid``` and ```AxesSubPlot```
  * Setting x and y labels
  * Rotating labels

# **Line plots**

**Overview**

* Visualization of the same "thing" through time
* Example: Tracking the value of a financial market stock over time

**Dataset**

* Air pollution data: Show nitrogen dioxide levels in air samples over time

````
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

mpg = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Course - Introduction to Data Visualization with Seaborn/mpg.csv")

#sns.relplot(x="model_year", y="mpg", kind="line", data=mpg, ci=None) # No confidence interval

#sns.relplot(x="model_year", y="mpg", kind="line", data=mpg) # Confidence interval is the mean

sns.relplot(x="model_year", y="mpg", 
            kind="line", data=mpg,
            ci="sd") # Confidence interval is the standard deviation

# Show plot
plt.show()

# The average miles per gallon has increased over time.
# We can be 95% confident that the average miles per gallon for all cars in 1970 is between 16 and 20 miles per gallon.
# This plot assumes that our data is a random sample of all cars in the US, Europe, and Japan
````

````
# Internet usage x age

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

survey_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Course - Introduction to Data Visualization with Seaborn/young-people-survey-responses.csv")

# Prepare columns "Age"
data = []
for index, d in enumerate(survey_data["Age"].values):
    if d < 21: data.append("Less than 21")
    else: data.append("21+")

df_data = pd.DataFrame({"Age":data})
survey_data["Age"] = df_data

# Make plot
sns.catplot(y="Internet usage", data=survey_data,
            kind="count", col="Age")
plt.show()
````

# **Box Plots**

# Overview
* Shows the distribution of quantitative data
* Show median, spread, skewness, outliers and confidence intervals
* Facilitates comparisons between groups
* It's encoraged to learn more about statistical analysis to understand the fundamentals of all the information that is been shown in the box plot.
  * Also, an statictical analysis course will help you edit the box plot accordingly to your needs.

# Usage
* Use of ```catplot```, with ```kind="box"```
* ```sym=""```: Ommit the outliers from the plot
* ```whis=2.0```: Make whiskers extend to 2.0 * IQR
* ```whis=[5, 95]```: Show the 5th and 95th percentiles
* ```whis=[0, 100]```: Show min and max values

````
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Creating dataset
student_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Course - Introduction to Data Visualization with Seaborn/student_data.csv", index_col=0)

# Preparing box plot
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]
sns.catplot(x="study_time", y="G3", 
            kind="box", 
            data=student_data, 
            order=study_time_order, 
            sym="") # Remove outliers
plt.show()
````

````
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

student_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Course - Introduction to Data Visualization with Seaborn/student_data.csv", index_col=0)

# Create a box plot with subgroups and omit the outliers
sns.catplot(x="internet", y="G3", 
            kind="box", 
            data=student_data, 
            hue="location"
            , sym=""   # Remove outliers
            #, whis=0.5 # Change whiskers from 1.5 to 0.5*IQR
            #, whis=[0, 100] # Change whiskers to min (0) and máx (100)
            ) 

# Show plot
plt.show()
````

# **Point plots**

## Overview
* Point plots have similiar information as box plots
* Points show mean of quantitative variable
* Vertical lines show 95% confidence intervals
 * Confidence intervals: show the level of uncertainty about the mean stimates.
 * Assuming that the data is a random sample of the population, we can be 95% sure that the true population mean in each group lies within the confidence interval shown.
* Main difference for point plots is that it has a **categorical** variabble on x-axis

## Usage
* ```catplot()``` with ```kind="point"```
* ```join=False```: Does not unite the points on the plot

````
# family relationship vs. absences

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

student_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Course - Introduction to Data Visualization with Seaborn/student_data.csv", index_col=0)
sns.catplot(x="famrel", y="absences", 
            kind="point", 
            data=student_data)
plt.show()
````

````
# Creating a point plot with subgroups

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

student_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Course - Introduction to Data Visualization with Seaborn/student_data.csv", index_col=0)
sns.catplot(x="romantic", y="absences", 
            kind="point", 
            data=student_data,
            hue="school", # Make plots with different colors
            ci=None, # Confidence interval is off
            estimator=np.median) # Display the median number of absences instead of the average
plt.show()
````

# **Customization**

* **Changing style**: ```sns.set_style()```
  * Options are "white", "dark", "whitegrid", "darkgrid", "ticks"
    * "ticks" show a tick for each value in the x and y-axis
* **Changing palette**: ```sns.set_palette()```
  * Mix colors: "RdBu", "PRGn", "RdBu_r" and "PRGn_r" 
  * Sequential colors: "Greys", "Blues", "PuRd" and "GnBu"
  * Custom palette with color names: 
    * ```custom_palette = ["red", "green","orange", "blue", "yellow", "purple"]```
  * Custom palette with hexadecimal: 
    * ```custom_palette = ['#FBB4AE', '#B3CDE3', '#CCEBC5', '#DECBE4', '#FED9A6', '#FFFFCC', '#E5D8BD', '#FDDAEC', '#F2F2F2']```  
  * **Changing scale**: ```sns.set_context()```
    * From smallest to largest: "paper", "notebook", "talk", "poster"


````
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

survey_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Course - Introduction to Data Visualization with Seaborn/young-people-survey-responses.csv")

sns.set_style("whitegrid")
#sns.set_style("darkgrid")

#sns.set_palette("Greens")
#sns.set_palette("Purples")

cp = ["#39A7D0", "#36ADA4"]
sns.set_palette(cp)

#sns.set_context("paper")
#sns.set_context("notebook")
sns.set_context("talk")
#sns.set_context("poster")

gender = ["male", "female"] # To set the order of the bars in the plot
sns.catplot(x="Gender", y="Mathematics", 
            data=survey_data, kind="bar", 
            order=gender)
plt.show()
````

## FacetGrid vs AxesSubplot objects
* ```FacetGrid``` is created for plot types ```relplot()``` and ```catplot()```, which can create **subplots**
* ```AxesSubplot``` is created for plot typles ```scatterplot()```, ```countplot()```, etc, ehich can only create a **single plot**


* These two values are returned when using ```sns.scatterplot()```, ```sns.relplot()```, etc, and can be stored in a variable
  *  ```g = sns.catplot(...)```

## FaceGrid customizations
* ```g.fig.suptitle("New Title", y=1.03)```: Specify title and height of the figure
